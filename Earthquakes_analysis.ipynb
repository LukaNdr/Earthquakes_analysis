```python
import numpy as np
import pandas as pd
df = pd.read_csv('Earthquakes_database.csv')
df.head()
```

```python
df = df.drop_duplicates()
df
```

```python
df = df.drop(columns = "ID")
df
```

```python
df[["Month", "day", "Year"]] = df["Date"].str.split('/',2, expand=True)
df
```

```python
bins = [0, 50, 150, float('inf')]
labels = ['Normal', 'Depth', 'Very depth']

df['Depth_Category'] = pd.cut(df['Depth'], bins=bins, labels=labels, right=False)

df.head()
```

```python
df_stacked=df.stack()
df_stacked
```

```python
df.melt()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))
df.groupby('Year')['Magnitude'].count().plot(kind='line', marker='o', color='blue')
plt.title('Earthquake Count Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Earthquakes')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Depth_Category', data=df, palette='viridis')
plt.title('Distribution of Earthquakes by Depth Category')
plt.xlabel('Depth Category')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(df['Longitude'], df['Latitude'], c='red', alpha=0.5, edgecolors='k', linewidths=0.5)
plt.title('Earthquake Distribution - Latitude vs Longitude')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```

```python
depth_category_counts = df['Depth_Category'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(depth_category_counts, labels=depth_category_counts.index, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightgreen', 'lightcoral'])
plt.title('Proportion of Earthquakes by Depth Category')
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'], utc=True)

print(df['Date'].dtype)

if df['Date'].dtype == 'datetime64[ns, UTC]':
    df['DayOfWeek'] = df['Date'].dt.day_name()

print(df[['Date', 'DayOfWeek']].head())
```

```python
df.set_index('Date', inplace=True)

# Resample to monthly frequency and calculate the mean magnitude
monthly_avg_magnitude = df['Magnitude'].resample('M').mean()

# Display the resampled data
print(monthly_avg_magnitude.head())
```
